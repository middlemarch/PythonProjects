### To understand a bit more about the leavers, we can use kmeans clustering, unsupervised- i.e. with no target categories. We start by choosing the continuous variables as this algorithm relies on distances to compute clusters.

```python
cluster_cols = ['satisfaction_level',  
                'last_evaluation', 
                'number_project', 
                'average_montly_hours', 
                'time_spend_company',
                ]
```

### Now to standardise the figures using sklearn's MaxAbsScaler, which will convert all values to a range between 0 and 1 (for positive values)

```python
scaler = preprocessing.MaxAbsScaler()
hr_clusters = scaler.fit_transform(df_left[cluster_cols])
hr_clusters = pd.DataFrame(hr_clusters, columns=cluster_cols)
```

### Some functions for preparing and summarising the clusters 
#### (credit: Tony Ojeda, http://blog.districtdatalabs.com/data-exploration-with-python-2)

```python
def kmeans_cluster(df, n_clusters=2):
    model = KMeans(n_clusters=n_clusters, random_state=1)
    clusters = model.fit_predict(df)
    cluster_results = df.copy()
    cluster_results['Cluster'] = clusters
    return cluster_results
```

```python
def summarize_clustering(results):
    cluster_size = results.groupby(['Cluster']).size().reset_index()
    cluster_size.columns = ['Cluster', 'Count']
    cluster_means = results.groupby(['Cluster'], as_index=False).mean()
    cluster_summary = pd.merge(cluster_size, cluster_means, on='Cluster')
    return cluster_summary
```

### After some tests, the appropriate number of clusters with distinct characteristics was 3- the magic number!

```python
cluster_results = kmeans_cluster(hr_clusters, 3)
cluster_summary = summarize_clustering(cluster_results)
```

```python
cluster_summary
```

### Again, a heatmap allows us to visualise the results

```python
sns.heatmap(cluster_summary[cluster_cols].transpose(), annot=True)
```

### Looking at the heatmap above, we appear to have 3 types of leavers:
* Hard-working people, highly regarded, with the company a while but not very happy. These people could be over-worked
* Pretty average people across the board; natural attrition perhaps
* Really happy, hard-working people with long service, likely among the most treasured employees. Maybe head-hunted or needing a new challenge?
